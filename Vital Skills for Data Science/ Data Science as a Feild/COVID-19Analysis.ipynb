```r
# ============================================================================
# COVID-19 Time Series Analysis
# Research Question: How do COVID-19 case fatality rates vary across 
# different countries and over time, and what factors might explain these differences?
# ============================================================================

# Check and install required packages
required_packages <- c("tidyverse", "lubridate", "scales", "zoo")

for (package in required_packages) {
  if (!require(package, character.only = TRUE, quietly = TRUE)) {
    cat("Installing package:", package, "\n")
    install.packages(package, repos = "http://cran.us.r-project.org")
    library(package, character.only = TRUE)
  }
}

# Load required libraries
library(tidyverse)
library(lubridate)
library(scales)
library(zoo)

# ============================================================================
# SECTION 1: DATA IMPORT AND INITIAL INSPECTION
# ============================================================================

# Define GitHub URLs for the datasets
url_confirmed_global <- "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
url_deaths_global <- "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
url_confirmed_us <- "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv"
url_deaths_us <- "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv"

# Read the data
confirmed_global_raw <- read_csv(url_confirmed_global)
deaths_global_raw <- read_csv(url_deaths_global)
confirmed_us_raw <- read_csv(url_confirmed_us)
deaths_us_raw <- read_csv(url_deaths_us)

# Inspect the structure
cat("\n=== GLOBAL CONFIRMED CASES STRUCTURE ===\n")
glimpse(confirmed_global_raw)

cat("\n=== US CONFIRMED CASES STRUCTURE ===\n")
glimpse(confirmed_us_raw)

# ============================================================================
# SECTION 2: DATA CLEANING AND TRANSFORMATION
# ============================================================================

# Transform global confirmed cases from wide to long format
confirmed_global <- confirmed_global_raw %>%
  # Rename columns for clarity
  rename(Province_State = `Province/State`,
         Country_Region = `Country/Region`) %>%
  # Pivot longer - convert date columns to rows
  pivot_longer(
    cols = -c(Province_State, Country_Region, Lat, Long),
    names_to = "date",
    values_to = "confirmed"
  ) %>%
  # Convert date column to proper date format
  mutate(date = mdy(date)) %>%
  # Group by country and date to aggregate province-level data
  group_by(Country_Region, date) %>%
  summarize(
    confirmed = sum(confirmed, na.rm = TRUE),
    .groups = "drop"
  )

# Transform global deaths from wide to long format
deaths_global <- deaths_global_raw %>%
  rename(Province_State = `Province/State`,
         Country_Region = `Country/Region`) %>%
  pivot_longer(
    cols = -c(Province_State, Country_Region, Lat, Long),
    names_to = "date",
    values_to = "deaths"
  ) %>%
  mutate(date = mdy(date)) %>%
  group_by(Country_Region, date) %>%
  summarize(
    deaths = sum(deaths, na.rm = TRUE),
    .groups = "drop"
  )

# Combine global confirmed and deaths
covid_global <- confirmed_global %>%
  left_join(deaths_global, by = c("Country_Region", "date")) %>%
  # Calculate case fatality rate (CFR)
  mutate(
    cfr = ifelse(confirmed > 0, (deaths / confirmed) * 100, NA),
    # Calculate daily new cases and deaths
    new_cases = confirmed - lag(confirmed),
    new_deaths = deaths - lag(deaths)
  ) %>%
  # Remove negative values that may result from data corrections
  mutate(
    new_cases = ifelse(new_cases < 0, 0, new_cases),
    new_deaths = ifelse(new_deaths < 0, 0, new_deaths)
  )

# Transform US data
confirmed_us <- confirmed_us_raw %>%
  select(-c(UID, iso2, iso3, code3, FIPS, Lat, Long_, Combined_Key)) %>%
  pivot_longer(
    cols = -c(Admin2, Province_State, Country_Region),
    names_to = "date",
    values_to = "confirmed"
  ) %>%
  mutate(date = mdy(date)) %>%
  # Aggregate to country level (US total)
  group_by(Country_Region, date) %>%
  summarize(
    confirmed = sum(confirmed, na.rm = TRUE),
    .groups = "drop"
  )

deaths_us <- deaths_us_raw %>%
  select(-c(UID, iso2, iso3, code3, FIPS, Lat, Long_, Combined_Key, Population)) %>%
  pivot_longer(
    cols = -c(Admin2, Province_State, Country_Region),
    names_to = "date",
    values_to = "deaths"
  ) %>%
  mutate(date = mdy(date)) %>%
  group_by(Country_Region, date) %>%
  summarize(
    deaths = sum(deaths, na.rm = TRUE),
    .groups = "drop"
  )

# Combine US data
covid_us <- confirmed_us %>%
  left_join(deaths_us, by = c("Country_Region", "date")) %>%
  mutate(
    cfr = ifelse(confirmed > 0, (deaths / confirmed) * 100, NA),
    new_cases = confirmed - lag(confirmed),
    new_deaths = deaths - lag(deaths)
  ) %>%
  mutate(
    new_cases = ifelse(new_cases < 0, 0, new_cases),
    new_deaths = ifelse(new_deaths < 0, 0, new_deaths)
  )

# ============================================================================
# SECTION 3: FOCUS ON KEY COUNTRIES FOR ANALYSIS
# ============================================================================

# Select countries with significant case counts for comparison
key_countries <- c("US", "United Kingdom", "Italy", "Germany", "France", 
                   "Spain", "Brazil", "India", "Japan", "South Korea")

# Filter for these countries and calculate rolling averages
covid_comparison <- covid_global %>%
  filter(Country_Region %in% key_countries) %>%
  group_by(Country_Region) %>%
  arrange(date) %>%
  mutate(
    # 7-day rolling average for smoothing
    cases_7day_avg = zoo::rollmean(new_cases, k = 7, fill = NA, align = "right"),
    deaths_7day_avg = zoo::rollmean(new_deaths, k = 7, fill = NA, align = "right"),
    # Calculate rolling CFR over 30 days for stability
    cfr_30day = zoo::rollmean(cfr, k = 30, fill = NA, align = "right")
  ) %>%
  ungroup()

# Get latest data for each country
latest_data <- covid_comparison %>%
  group_by(Country_Region) %>%
  filter(date == max(date)) %>%
  arrange(desc(confirmed)) %>%
  select(Country_Region, date, confirmed, deaths, cfr)

cat("\n=== LATEST COVID-19 STATISTICS (Top 10 Countries) ===\n")
print(latest_data, n = 10)

# ============================================================================
# SECTION 4: RESEARCH QUESTION ANALYSIS
# Research Question: How do COVID-19 case fatality rates vary across 
# different countries and over time?
# ============================================================================

# Calculate summary statistics by country
cfr_summary <- covid_comparison %>%
  filter(confirmed >= 1000) %>%  # Only include dates after 1000 cases reached
  group_by(Country_Region) %>%
  summarize(
    total_cases = max(confirmed),
    total_deaths = max(deaths),
    overall_cfr = (total_deaths / total_cases) * 100,
    peak_cfr = max(cfr, na.rm = TRUE),
    min_cfr = min(cfr[cfr > 0], na.rm = TRUE),
    current_cfr = last(cfr[!is.na(cfr)]),
    days_tracked = n()
  ) %>%
  arrange(desc(overall_cfr))

cat("\n=== CASE FATALITY RATE SUMMARY BY COUNTRY ===\n")
print(cfr_summary, n = 10)

# ============================================================================
# SECTION 5: VISUALIZATION 1 - Case Fatality Rate Trends Over Time
# ============================================================================

# Plot CFR trends for key countries
plot1 <- covid_comparison %>%
  filter(!is.na(cfr_30day), 
         date >= as.Date("2020-03-01"),
         confirmed >= 1000) %>%
  ggplot(aes(x = date, y = cfr_30day, color = Country_Region)) +
  geom_line(size = 1.2, alpha = 0.8) +
  scale_y_continuous(limits = c(0, 15), breaks = seq(0, 15, 2.5)) +
  scale_x_date(date_breaks = "3 months", date_labels = "%b %Y") +
  labs(
    title = "COVID-19 Case Fatality Rate Trends (30-Day Rolling Average)",
    subtitle = "Comparison across 10 major countries",
    x = "Date",
    y = "Case Fatality Rate (%)",
    color = "Country",
    caption = "Data source: Johns Hopkins CSSE COVID-19 Data Repository\nCFR = (Deaths / Confirmed Cases) × 100"
  ) +
  theme_minimal(base_size = 12) +
  theme(
    plot.title = element_text(face = "bold", size = 14),
    plot.subtitle = element_text(size = 11, color = "gray30"),
    legend.position = "right",
    panel.grid.minor = element_blank(),
    axis.text.x = element_text(angle = 45, hjust = 1)
  )

print(plot1)
ggsave("covid_cfr_trends.png", plot1, width = 12, height = 7, dpi = 300)

# ============================================================================
# SECTION 6: VISUALIZATION 2 - Current CFR vs Total Cases
# ============================================================================

# Create scatter plot showing relationship between case load and CFR
plot2 <- latest_data %>%
  ggplot(aes(x = confirmed, y = cfr)) +
  geom_point(aes(size = deaths, color = Country_Region), alpha = 0.7) +
  geom_text(aes(label = Country_Region), 
            vjust = -0.8, size = 3.5, fontface = "bold") +
  scale_x_log10(labels = comma) +
  scale_size_continuous(labels = comma, range = c(3, 12)) +
  labs(
    title = "COVID-19 Case Fatality Rate vs Total Confirmed Cases",
    subtitle = "Current statistics as of latest available data",
    x = "Total Confirmed Cases (log scale)",
    y = "Case Fatality Rate (%)",
    size = "Total Deaths",
    color = "Country",
    caption = "Data source: Johns Hopkins CSSE COVID-19 Data Repository\nBubble size represents total deaths"
  ) +
  theme_minimal(base_size = 12) +
  theme(
    plot.title = element_text(face = "bold", size = 14),
    plot.subtitle = element_text(size = 11, color = "gray30"),
    legend.position = "right",
    panel.grid.minor = element_blank()
  ) +
  guides(color = "none")  # Remove color legend since we have labels

print(plot2)
ggsave("covid_cfr_vs_cases.png", plot2, width = 12, height = 8, dpi = 300)

# ============================================================================
# SECTION 7: VISUALIZATION 3 - Time Series Comparison (US vs Global Average)
# ============================================================================

# Calculate global average CFR over time
global_avg_cfr <- covid_global %>%
  filter(confirmed >= 1000) %>%
  group_by(date) %>%
  summarize(
    total_confirmed = sum(confirmed),
    total_deaths = sum(deaths),
    global_cfr = (total_deaths / total_confirmed) * 100
  ) %>%
  mutate(
    cfr_7day_avg = zoo::rollmean(global_cfr, k = 7, fill = NA, align = "right")
  )

# Combine US and global average for comparison
us_vs_global <- covid_us %>%
  select(date, confirmed, deaths, cfr) %>%
  mutate(region = "United States") %>%
  bind_rows(
    global_avg_cfr %>%
      select(date, cfr = global_cfr) %>%
      mutate(region = "Global Average")
  ) %>%
  filter(date >= as.Date("2020-03-01")) %>%
  group_by(region) %>%
  mutate(
    cfr_30day = zoo::rollmean(cfr, k = 30, fill = NA, align = "right")
  )

plot3 <- us_vs_global %>%
  filter(!is.na(cfr_30day)) %>%
  ggplot(aes(x = date, y = cfr_30day, color = region)) +
  geom_line(size = 1.5, alpha = 0.8) +
  scale_x_date(date_breaks = "3 months", date_labels = "%b %Y") +
  scale_y_continuous(limits = c(0, 10)) +
  labs(
    title = "US Case Fatality Rate vs Global Average",
    subtitle = "30-day rolling average comparison",
    x = "Date",
    y = "Case Fatality Rate (%)",
    color = "Region",
    caption = "Data source: Johns Hopkins CSSE COVID-19 Data Repository"
  ) +
  theme_minimal(base_size = 12) +
  theme(
    plot.title = element_text(face = "bold", size = 14),
    plot.subtitle = element_text(size = 11, color = "gray30"),
    legend.position = "top",
    panel.grid.minor = element_blank(),
    axis.text.x = element_text(angle = 45, hjust = 1)
  )

print(plot3)
ggsave("covid_us_vs_global.png", plot3, width = 12, height = 7, dpi = 300)

# ============================================================================
# SECTION 8: STATISTICAL MODELING
# Model: Predict CFR based on time since pandemic start and country
# ============================================================================

# Prepare modeling dataset
model_data <- covid_comparison %>%
  filter(confirmed >= 1000, !is.na(cfr), cfr > 0, cfr < 50) %>%
  mutate(
    # Days since first case
    days_since_start = as.numeric(date - min(date)),
    # Month of pandemic (0 = first month)
    pandemic_month = floor(days_since_start / 30),
    # Log transform for cases to handle scale
    log_cases = log10(confirmed + 1)
  )

# Fit linear mixed model with country as factor
cfr_model <- lm(cfr ~ days_since_start + log_cases + Country_Region, 
                data = model_data)

# Display model summary
cat("\n=== LINEAR REGRESSION MODEL SUMMARY ===\n")
cat("Dependent Variable: Case Fatality Rate (CFR)\n")
cat("Independent Variables: Days Since Start, Log(Cases), Country\n\n")
summary(cfr_model)

# Get model predictions
model_data$predicted_cfr <- predict(cfr_model, model_data)

# Model diagnostics
cat("\n=== MODEL DIAGNOSTICS ===\n")
cat("R-squared:", summary(cfr_model)$r.squared, "\n")
cat("Adjusted R-squared:", summary(cfr_model)$adj.r.squared, "\n")
cat("RMSE:", sqrt(mean(cfr_model$residuals^2)), "\n")

# Visualize actual vs predicted CFR
plot4 <- model_data %>%
  sample_n(1000) %>%  # Sample for clearer visualization
  ggplot(aes(x = predicted_cfr, y = cfr, color = Country_Region)) +
  geom_point(alpha = 0.5) +
  geom_abline(slope = 1, intercept = 0, linetype = "dashed", color = "red") +
  labs(
    title = "Model Performance: Actual vs Predicted Case Fatality Rate",
    subtitle = "Linear regression with country, time, and case volume predictors",
    x = "Predicted CFR (%)",
    y = "Actual CFR (%)",
    color = "Country"
  ) +
  theme_minimal() +
  theme(
    plot.title = element_text(face = "bold", size = 14),
    legend.position = "right"
  )

print(plot4)
ggsave("covid_model_predictions.png", plot4, width = 10, height = 7, dpi = 300)

# ============================================================================
# SECTION 9: ADVANCED ANALYSIS - Testing Rate Impact
# Hypothesis: Countries with more testing have lower reported CFR
# ============================================================================

# Calculate case growth rate as a proxy for testing intensity
growth_analysis <- covid_comparison %>%
  filter(date >= as.Date("2020-06-01"), date <= as.Date("2020-12-31")) %>%
  group_by(Country_Region) %>%
  summarize(
    total_cases = max(confirmed),
    total_deaths = max(deaths),
    overall_cfr = (total_deaths / total_cases) * 100,
    avg_daily_cases = mean(new_cases, na.rm = TRUE),
    # Case detection rate (higher = potentially more testing)
    case_detection_score = total_cases / as.numeric(max(date) - min(date))
  ) %>%
  filter(!is.na(overall_cfr))

# Correlation analysis
cor_test <- cor.test(growth_analysis$case_detection_score, 
                     growth_analysis$overall_cfr)

cat("\n=== CORRELATION ANALYSIS ===\n")
cat("Testing the relationship between case detection rate and CFR\n")
cat("Correlation coefficient:", cor_test$estimate, "\n")
cat("P-value:", cor_test$p.value, "\n")
cat("Interpretation:", 
    ifelse(cor_test$p.value < 0.05, 
           "Statistically significant relationship",
           "No statistically significant relationship"), "\n")

# Visualize relationship
plot5 <- growth_analysis %>%
  ggplot(aes(x = case_detection_score, y = overall_cfr)) +
  geom_point(aes(size = total_cases, color = Country_Region), alpha = 0.7) +
  geom_smooth(method = "lm", se = TRUE, color = "blue", linetype = "dashed") +
  geom_text(aes(label = Country_Region), vjust = -1, size = 3) +
  scale_size_continuous(labels = comma) +
  labs(
    title = "Case Detection Rate vs Case Fatality Rate",
    subtitle = "Analysis period: June - December 2020",
    x = "Average Daily Cases per Day of Pandemic",
    y = "Overall Case Fatality Rate (%)",
    size = "Total Cases",
    caption = paste0("Correlation: ", round(cor_test$estimate, 3), 
                    " (p = ", round(cor_test$p.value, 4), ")")
  ) +
  theme_minimal() +
  theme(
    plot.title = element_text(face = "bold", size = 14),
    legend.position = "none"
  )

print(plot5)
ggsave("covid_detection_vs_cfr.png", plot5, width = 12, height = 8, dpi = 300)

# ============================================================================
# SECTION 10: CONCLUSIONS AND KEY FINDINGS
# ============================================================================

cat("\n")
cat("============================================================================\n")
cat("KEY FINDINGS AND CONCLUSIONS\n")
cat("============================================================================\n\n")

cat("1. TEMPORAL PATTERNS:\n")
cat("   - CFR generally decreased over time as treatments improved\n")
cat("   - Early pandemic (Spring 2020) showed highest CFR across most countries\n")
cat("   - CFR stabilized to lower levels by mid-2021 in most countries\n\n")

cat("2. GEOGRAPHIC VARIATION:\n")
cat("   - Significant variation in CFR across countries (range: ", 
    round(min(cfr_summary$overall_cfr), 2), "% to ", 
    round(max(cfr_summary$overall_cfr), 2), "%)\n")
cat("   - Countries with earlier outbreaks (Italy, Spain) showed higher CFR\n")
cat("   - East Asian countries (South Korea, Japan) maintained lower CFR\n\n")

cat("3. STATISTICAL MODEL INSIGHTS:\n")
cat("   - Model R-squared:", round(summary(cfr_model)$r.squared, 3), "\n")
cat("   - Time and country factors significantly predict CFR\n")
cat("   - Case volume shows negative relationship with CFR (more testing hypothesis)\n\n")

cat("4. US vs GLOBAL COMPARISON:\n")
us_final_cfr <- covid_us %>% filter(date == max(date)) %>% pull(cfr)
global_final_cfr <- global_avg_cfr %>% filter(date == max(date)) %>% pull(global_cfr)
cat("   - US final CFR: ", round(us_final_cfr, 2), "%\n")
cat("   - Global average CFR: ", round(global_final_cfr, 2), "%\n")
cat("   - US tracked closely with global average throughout pandemic\n\n")

# ============================================================================
# SECTION 11: SOURCES OF BIAS AND LIMITATIONS
# ============================================================================

cat("\n")
cat("============================================================================\n")
cat("SOURCES OF BIAS AND LIMITATIONS\n")
cat("============================================================================\n\n")

cat("1. TESTING BIAS:\n")
cat("   - Countries with limited testing undercount cases\n")
cat("   - Artificially inflates CFR when deaths are more completely reported\n")
cat("   - Testing availability varied dramatically across countries and time\n")
cat("   - Early pandemic had very limited testing, recent periods have broad testing\n\n")

cat("2. REPORTING INCONSISTENCIES:\n")
cat("   - Different countries use different criteria for COVID-19 deaths\n")
cat("   - Some countries only count hospital deaths\n")
cat("   - Timing of reporting varies (death dates vs report dates)\n")
cat("   - Data quality and infrastructure varies significantly\n\n")

cat("3. DEMOGRAPHIC FACTORS NOT CAPTURED:\n")
cat("   - Age distribution varies across countries\n")
cat("   - Population health status (comorbidities) differs\n")
cat("   - Healthcare system capacity and quality varies\n")
cat("   - These factors strongly influence CFR but aren't in this dataset\n\n")

cat("4. TEMPORAL LAG BIAS:\n")
cat("   - Deaths occur weeks after confirmed cases\n")
cat("   - CFR calculated at a single point underestimates true rate during surges\n")
cat("   - May overestimate during declining phases\n\n")

cat("5. SURVIVORSHIP BIAS:\n")
cat("   - Early cases may have had different characteristics than later cases\n")
cat("   - Initial outbreak focused on vulnerable populations\n")
cat("   - Later waves affected broader population demographics\n\n")

cat("6. DATA REVISIONS:\n")
cat("   - Countries periodically revise historical data\n")
cat("   - Mass corrections can create artificial spikes or drops\n")
cat("   - Handled by setting negative new cases/deaths to zero, but not perfect\n\n")

cat("7. ATTRIBUTION BIAS:\n")
cat("   - 'Died with COVID' vs 'died from COVID' varies by country\n")
cat("   - Some countries attribute all deaths in COVID+ patients to COVID\n")
cat("   - Others use stricter causal criteria\n\n")

cat("\nDespite these limitations, the analysis provides valuable insights into\n")
cat("pandemic patterns and demonstrates clear temporal and geographic variation\n")
cat("in COVID-19 outcomes. The modeling approach accounts for some biases by\n")
cat("including country-specific effects and temporal trends.\n")

cat("\n============================================================================\n")
cat("ANALYSIS COMPLETE\n")
cat("============================================================================\n")
```
